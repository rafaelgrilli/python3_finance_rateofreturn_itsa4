```python
# on our columns' name, we have special characters such as parenthesis, & and others. We may want to replace it 
# with underscores, and the following function does the conversion and use a loop apply this change automatically.
```

```python
def get_better_col_name(col):
    #replace special symbols
    special_symbols = ['&', ',','.','(', ')']
    
    for s in special_symbols: 
        col = col.replace(s, '')
# this first bloc of code removes special symbols, looping through the dataset one by one.  


    # as a good practice, put all the columns' names in capital letters, without any space around the name
    col = col.upper().strip()
    
    # split the string into list by space
    col_split = col.split(' ')
    
    # remove the extra space by going through col_split
    # then append the part with actual letters into col_split_clean
    col_split_clean = []
    
    for part in col_split:
        if not part.strip() == ' ':
            col_split_clean.append(part.strip())
# to get rid of the extra space generated when special symbols were taken out, we create an empty list 
# "col_split_clean" and loop through the original list "col_split". 
            
            
    # connect the list by string "_" into one word
    col_new = '_'.join(col_split_clean)
# we use the funcion "join" to connect all the workds in the list "col_split_clean" with underscore "_"
    
    return col_new

get_better_col_name('Non-Operating Income (Loss)')
```

```python
# on the next loop we use the prior function to prepare a dictionary (map) and a list of new column headers
```

```python
col_map = {}
col_new_list = []
# we created an empty dictionary and a list for the loop.

col_list = df.columns.tolist()

for col in col_list: 
    col_new = get_better_col_name(col)
    col_map[col] = col_new
    col_new_list.append(col_new)
# within the loop, we called the function and converted the old column headers into new ones; 
# for a dictionary, we created the mapping, old column header as the key and the new column header as the value;

    
# print the list
print(col_new_list)

# let jupyter display the dictionary
col_map
```

```python
df_better_name = df.rename(columns=col_map)
df_better_name.columns.tolist()
```

```python
# computing financial ratios. first, copy the sub data set and assign back to the variable "df_profit_2019_q3"
```

```python
df_profit_2019 = df_profit_2019.copy()
```

```python
df_profit_2019['EPS'] = df_profit_2019.NET_INCOME / df_profit_2019.SHARES_BASIC 
df_profit_2019['GROSS_PROFIT_MARGIN'] = df_profit_2019.GROSS_PROFIT / df_profit_2019.REVENUE * 100
df_profit_2019['PROFIT_MARGIN'] = df_profit_2019.NET_INCOME / df_profit_2019.REVENUE * 100

df_profit_2019[['TICKER', 'EPS', 'GROSS_PROFIT_MARGIN', 'PROFIT_MARGIN']].head()
```

```python
# now let's say that we want to tag the company "above average" when it has eps larger than the average,
# otherwise, we tag it "below average". 
```

```python
df_profit_2019.EPS.describe()
```

```python
def classify_eps(eps):
    eps_class = 'Below Average'
    
    if eps > 3.070920:
        eps_class = 'Above Average'
        
    return eps_class
```

```python
df_profit_2019['EPS_CLASS'] = df_profit_2019.EPS.apply(classify_eps)

df_profit_2019[['TICKER', 'EPS_CLASS']].head(20)
```

```python
# decile rank scores = we want to score three factors: EPS, GROSS Margin and Net Profit Margin. 
# one common way is the decile rank, sort the companies by given factor and evenly slice then into 10 groups.
# the score will be between 1 to 10 and 10 is the best. we use the function "qcut"
# this is the reason for the "10". for percentile or quantile, we can put 100 or 5 respectively. 
```

```python
df_profit_2019['EPS_SCORE'] = pd.qcut(df_profit_2019.EPS, 10, labels=False)
df_profit_2019['GROSS_PROFIT_MARGIN_SCORE'] = pd.qcut(df_profit_2019.NET_INCOME, 10, labels=False)
df_profit_2019['PROFIT_MARGIN_SCORE'] = pd.qcut(df_profit_2019.PROFIT_MARGIN, 10, labels=False)

df_profit_2019[['TICKER', 'EPS_SCORE', 'GROSS_PROFIT_MARGIN_SCORE', 'PROFIT_MARGIN_SCORE']].head()
```

```python
# the next lines of code calculate the average scores of profitability, then "sort_values" and "head"
# will help us to get the top 10
```

```python
df_profit_2019['AVG_SCORE'] = (
    df_profit_2019.EPS_SCORE + 
    df_profit_2019.GROSS_PROFIT_MARGIN_SCORE +
    df_profit_2019.PROFIT_MARGIN_SCORE
) / 3

df_profit_2019.sort_values('AVG_SCORE', ascending=False).head(10)[['TICKER', 'AVG_SCORE']]
```

```python
# on the next lines, we tag the company "Above Average" when it has all three-factor scores more than 4
# otherwise, we consider it as "Below Average". this is a quality measure.
```

```python
def classify_company(row):
    company_class = 'Below Average'
    
    if row['EPS_SCORE'] > 4 and row['GROSS_PROFIT_MARGIN_SCORE'] > 4 and row['PROFIT_MARGIN_SCORE'] > 4:
        company_class = 'Above Average'
    
    return company_class
```

```python
df_profit_2019['COMPANY_CLASS'] = df_profit_2019.apply(classify_company, axis=1)

# as a default axis=1 means we want to apply the function row by row. as a default, axis=0 is column by column. 

df_profit_2019[['TICKER', 'COMPANY_CLASS']].head(100)
```

```python
# applying z-score to measure how good/bad a factor is relative to the average level.
# the value of z-score means the number of std dev away from the mean.
# positive z-score means the company is better than the average and negative means the opposite. 
```

```python
eps_mean = df_profit_2019.EPS.mean()
eps_std = df_profit_2019.EPS.std()

print(eps_mean, eps_std)
```

```python
# we then do the math to calculate the z-scores:
df_profit_2019['EPS_Z_SCORE'] = (df_profit_2019.EPS - eps_mean) / eps_std

df_profit_2019[['TICKER', 'EPS_Z_SCORE']].head()
```

```python
# similarly, we can apply the same to all three factors, then calculate the average z-scores 
# to rank the companies and find the top 10
```

```python
eps_mean = df_profit_2019.EPS.mean()
eps_std = df_profit_2019.EPS.std()

df_profit_2019['EPS_Z_SCORE'] = (
    df_profit_2019.EPS - eps_mean
) / eps_std

net_income_mean = df_profit_2019.NET_INCOME.mean()
net_income_std = df_profit_2019.NET_INCOME.std()

df_profit_2019['NET_INCOME_Z_SCORE'] = (
    df_profit_2019.NET_INCOME - net_income_mean
) / net_income_std

profit_margin_mean = df_profit_2019.PROFIT_MARGIN.mean()
profit_margin_std = df_profit_2019.PROFIT_MARGIN.std()
df_profit_2019['PROFIT_MARGIN_Z_SCORE'] = (
    df_profit_2019.PROFIT_MARGIN - profit_margin_mean
) / profit_margin_std

df_profit_2019['AVG_Z_SCORE'] = (
    df_profit_2019.EPS_Z_SCORE +
    df_profit_2019.NET_INCOME_Z_SCORE + 
    df_profit_2019.PROFIT_MARGIN_Z_SCORE
) / 3

df_profit_2019.sort_values('AVG_Z_SCORE', ascending=False).head(10)[['TICKER', 'AVG_Z_SCORE']]
```

```python
# since z-score can potentially output extreme values, practioners usually adjust it with a 
# process called winsorize, which means cap the max and min values to a given level
```

```python
eps_mean = df_profit_2019.EPS.mean()
eps_std = df_profit_2019.EPS.std()

df_profit_2019['EPS_Z_SCORE'] - (
    df_profit_2019.EPS - eps_mean
) / eps_std

df_profit_2019.EPS_Z_SCORE = df_profit_2019.EPS_Z_SCORE.clip(-2,2)

net_income_mean = df_profit_2019.NET_INCOME.mean()
net_income_std = df_profit_2019.NET_INCOME.std()

df_profit_2019['NET_INCOME_Z_SCORE'] - (
    df_profit_2019.NET_INCOME - net_income_mean
) / net_income_std

df_profit_2019.NET_INCOME_Z_SCORE = df_profit_2019.NET_INCOME_Z_SCORE.clip(-2,2)

profit_margin_mean = df_profit_2019.PROFIT_MARGIN.mean()
profit_margin_std = df_profit_2019.PROFIT_MARGIN.std()

df_profit_2019['PROFIT_MARGIN_Z_SCORE'] - (
    df_profit_2019.PROFIT_MARGIN - profit_margin_mean
) / profit_margin_std

df_profit_2019.PROFIT_MARGIN_Z_SCORE = df_profit_2019.PROFIT_MARGIN_Z_SCORE.clip(-2,2)

df_profit_2019['AVG_Z_SCORE'] = (
    df_profit_2019.EPS_Z_SCORE +
    df_profit_2019.NET_INCOME_Z_SCORE + 
    df_profit_2019.PROFIT_MARGIN_Z_SCORE
) / 3

df_profit_2019.sort_values('AVG_Z_SCORE', ascending=False).head(10)[['TICKER', 'AVG_Z_SCORE']]
```

```python
df_industry = pd.read_csv('industries.csv', sep=';')
df_industry.head()
```

```python
df_company_info = pd.read_csv('us-companies.csv', sep=';')
df_company_info.head()
```

```python
# both above dataframes present one column bringing the same classification, "industryId'
# to make those data useful for the sector-neutral model, we need to "merge" the two dataframes 
# for given columns in commom - like Excel VLookup, but instead of taking one value, it connects all 
# associated columns in one go
```

```python
df_company_info_all = pd.merge(
    left=df_company_info,
    right=df_industry,
    how='left',
    on='IndustryId'
)
df_company_info_all.head()
```

```python
# pandas' merge function takes four inputs: left: the main DataFrame; right: the lookup DataFrame;
# how: lookup method; on: column/columns used for lookup
```

```python
df_company_info_all = df_company_info_all[['Ticker', 'Sector']]
df_company_info_all.columns = ['TICKER', 'SECTOR']
df_company_info_all.head()
```

```python
df_profit_2019_all = pd.merge(
    left=df_profit_2019,
    right=df_company_info_all,
    how='left', on='TICKER'
)

df_profit_2019_all.columns
```

```python
# now let's create a new DataFrame with all the data we need for the sector-neutral model.
```

```python
df_profit_sector = df_profit_2019_all[[
    'TICKER', 'SECTOR', 'EPS', 'NET_INCOME', 'PROFIT_MARGIN'
]]

df_profit_sector.head()
```

```python
# the sector-neutral model computes the z-scores only within the same sector. we could filter for a given
# sector and compute the scores, then loop through a list of unique sectors and merge the results. 
# with Pandas, we can use the function "groupby" to compute the sector statistics straight away.
```

```python
df_profit_sector_group = df_profit_sector.groupby('SECTOR')

df_profit_sector_group.count()
```

```python
# we will drop the sectors Energy & Business Services because they have only two stocks or fewer. 
```

```python
sector_cond = ~df_profit_sector.SECTOR.isin([
    'Energy', 'Business Services'
])

df_profit_sector_sub = df_profit_sector[sector_cond].copy()
df_profit_sector_sub.head()
```

```python
df_profit_sector_mean = df_profit_sector_sub.groupby('SECTOR').mean()
df_profit_sector_std = df_profit_sector_sub.groupby('SECTOR').std()
```

```python
df_profit_sector_mean
```

```python
df_profit_sector_std
```

```python
# now, let's compute the sector Z scores. 
```

```python
df_profit_sector_mean = df_profit_sector_mean.reset_index()
df_profit_sector_std = df_profit_sector_std.reset_index()
```

```python
df_profit_sector_mean
```

```python
df_profit_sector_std
```

```python
df_profit_sector_sub = pd.merge(
    left=df_profit_sector_sub,
    right=df_profit_sector_mean,
    how='left', 
    on='SECTOR',
    suffixes = ['', '_MEAN']
)
```

```python
df_profit_sector_sub = pd.merge(
    left=df_profit_sector_sub,
    right=df_profit_sector_std,
    how='left',
    on='SECTOR', 
    suffixes = ['', '_STD']
)
```

```python
df_profit_sector_sub.head()
```

```python
df_profit_sector_sub.columns
```

```python
# we now have all the three sector level statistics joined back to the stock level. we can then calculate
# the sector Z scores for each factor - EPS, Net Income and Profit Margin,
# and calculate the average z acors for ranking
```

```python
df_profit_sector_sub['EPS_Z'] = (
    df_profit_sector_sub.EPS - df_profit_sector_sub.EPS_MEAN
) / df_profit_sector_sub.EPS_STD

df_profit_sector_sub['NET_INCOME_Z'] = (
    df_profit_sector_sub.NET_INCOME - df_profit_sector_sub.NET_INCOME_MEAN
) / df_profit_sector_sub.NET_INCOME_STD

df_profit_sector_sub['PROFIT_MARGIN_Z'] = (
    df_profit_sector_sub.PROFIT_MARGIN - df_profit_sector_sub.PROFIT_MARGIN_MEAN
) / df_profit_sector_sub.PROFIT_MARGIN_STD

df_profit_sector_sub['Z'] = (
    df_profit_sector_sub.EPS_Z + 
    df_profit_sector_sub.NET_INCOME_Z + 
    df_profit_sector_sub.PROFIT_MARGIN_Z
) / 3
```

```python
df_top_10 = df_profit_sector_sub.sort_values('Z', ascending=False).head(10)
df_top_10
```
