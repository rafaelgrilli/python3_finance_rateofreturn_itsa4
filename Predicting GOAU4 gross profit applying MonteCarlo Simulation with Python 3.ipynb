#### In this exercise, we forecast Gerdau's (GOAU4) gross profit applying MonteCarlo Simulation. The assumptions adopted are: Revenues forecasted for FY 2020, end of the year = 41 bi (BR reais). Standart deviation of revenues = 3.5 bi (BR reais). COGs = 89% of revenues, with standart deviation of 1.5.

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
goau4_rev_M = 41
goau4_rev_stdev = 3.5
iterations = 10000
```

```python
rev = np.random.normal(goau4_rev_M, goau4_rev_stdev, iterations)
rev
```

```python
COGS = - (rev * np.random.normal(0.89, 0.015))
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
Gross_profit = rev + COGS
```

```python
Gross_profit
```

```python
plt.figure(figsize=(15,6))
plt.plot(Gross_profit)
plt.show()
```
