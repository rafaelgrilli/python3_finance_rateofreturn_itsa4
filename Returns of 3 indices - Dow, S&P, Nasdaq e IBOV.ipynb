# in this simple code, we estimate the returns of 3 American known indices, Dow, S&P and Nasdak, and Ibovespa, which is a nice proxy for the Brazilian stock market. 

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt 
```

```python
ind_data = pd.read_csv('Indices_data.csv', index_col='Date')
```

```python
ind_data.info()
```

```python
ind_data.tail()
```

# on the next line of the code we will normalize data to 100 and plot the results on a graph, which is useful for visually checking the indices daily behavior in the past 5 years. 

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(16,6));
plt.show()
```

# on the next lines of code, we calculate the simple returns for the indices 

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1

ind_returns.tail()
```

# on the next line of code we simply annualize the returns for the indices, which raw data was on a daily basis, as provided by YahooFinance. 

```python
ind_returns_a = ind_returns.mean() * 250
 
ind_returns_a
```
