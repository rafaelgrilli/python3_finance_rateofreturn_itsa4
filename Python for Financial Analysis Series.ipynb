```python
spy_df = spy_df[['open', 'high', 'low', 'close', 'volume']].copy()
spy_df = spy_df.sort_index()
spy_df.head()
```

```python
spy_df['pct_chg_daily'] = spy_df.close.pct_change()
spy_df.head()
```

```python
# above we took the column "close" and then apply the function "pct_change", 
# used to create the column "pct_chg_daily"
```

```python
spy_df.dropna().head()
```

```python
# in order to calculate log returns of each number and then calculate the net change, we need another
# package, Numpy, which provides mathematical and statistical functions
```

```python
import numpy as np
```

```python
spy_df['log_return'] = np.log(spy_df.close).diff() #like "pct_change", "diff" starts from the second row
# and calculate the net change on a rolling basis. 
spy_df.dropna().head()
```

```python
spy_df['close_5d'] = spy_df.close.shift(4) # 5 days % change
spy_df.head()
```

```python
spy_df['pct_chg_5d'] = spy_df.close / spy_df.close_5d - 1
spy_df.dropna().head()
```

```python
spy_df['SMA5'] = spy_df.close.rolling(5).mean()
spy_df.dropna().head()
```

```python
# the function "rolling" works with the function "apply" for more advanced computations. for instance,
# what if we would like to run a liner regression on the rolling basis and get the slope of the regression line?
# the rolling slope?
```

```python
def regression(y):
    return np.polyfit(range(len(y)), y, 1)[0] # numpy's "np.ployfit" function does the regression
# three parameters needed: "x", "y" and "degree" (with value 1, it does linear regression)

spy_df['slope30'] = spy_df.close.rolling(30).apply(regression, raw=True)
spy_df.dropna().head()
```

```python
# on the next lines, we will learn how to compute correlations, and more spefic rolling correlations. 
# we start by downloading a time series data for "IBM"
```

```python
ibm_df = c.chartDF('IBM', timeframe='5y')
ibm_df = ibm_df[['open', 'high', 'low', 'close', 'volume']].copy()
ibm_df = ibm_df.sort_index()
ibm_df.head()
```

```python
# now we join the column "close" from both DataFrames, IBM and SPY index
```

```python
data = spy_df[['close']].join(
    ibm_df[['close']], how='inner', lsuffix='_spy', rsuffix='_ibm'
)
data.head()
```

```python
# now we can compute the log return series based on the joined DataFrame and keep the log returns columns only.
```

```python
data['ret_spy'] = np.log(data.close_spy).diff()
data['ret_ibm'] = np.log(data.close_ibm).diff()

data = data.dropna()[['ret_spy', 'ret_ibm']]

data.head()
```

```python
data.corr()
```

```python
data.ret_spy.corr(data.ret_ibm) # for a single correlation number as the output
```

```python
# on the next line we calculate the rolling 30 days correlation between ibm and spy
```

```python
data.ret_spy.rolling(30).corr(data.ret_ibm).dropna().head()
```
